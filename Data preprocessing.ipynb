## Data preprocessing

Importamos librer√≠as

```python
import numpy as np
import pandas as pd
import ast
```

Importamos los datasets

```python
games = pd.read_csv('datasets/Games.csv')
```

# Dataset games

```python
games.head()
```

Quitamos las columnas que no se necesitan para el proyecto

```python
games = games[['publisher', 'genres', 'app_name', 'tags', 'discount_price', 'specs', 'price', 'early_access', 'id', 'developer', 'metascore', 'release_date']]
games.head()
```

Borramos los registros nulos

```python
games = games.dropna(how='all')
games.head()
```

Borramos los registros que tienen nulos en las categor√≠as: publisher, genres, app_name, tags, discount_price, specs, price, early_access, id y developer

```python
games = games.dropna(subset=['publisher', 'genres', 'app_name', 'tags', 'discount_price', 'specs', 'price', 'early_access', 'id', 'developer'])
```

```python
games.head()
```

Creamos la columna year para que solo nos muestre el a√±o de lanzamiento

```python
games['year'] = pd.to_datetime(games['release_date']).dt.year
```

```python
games.drop(columns='release_date', inplace=True)
```

```python
games.head()
```

Borramos registros duplicados

```python
games.drop_duplicates(inplace=True)
```

Nos aseguramos que las columnas que contienen listas no est√©n en formato str

```python
games['genres'] = games['genres'].apply(ast.literal_eval)
games['tags'] = games['tags'].apply(ast.literal_eval)
games['specs'] = games['specs'].apply(ast.literal_eval)
```

Guardamos el dataset limpio en formato csv

```python
games.to_csv('games_cleaned.csv', index=False)
```

# Dataset reviews

Cargamos el archivo

```python
reviews = pd.read_csv('datasets/reviews.csv')
```

Borramos los duplicados respecto a la columna user_id

```python
reviews.drop_duplicates(subset='user_id', inplace=True)
```

```python
reviews.head()
```

Creamos un dataset donde vamos a poner cada rese√±a para que sea m√°s f√°cil hacer el sentiment analysis

```python
review_singular = pd.DataFrame(columns=['user_id', 'item_id', 'review', 'recommend'])
```

```python
review_singular['item_id'] = pd.Series([])
```

```python
review_singular.head()
```

```python
reviews['reviews'] = reviews['reviews'].apply(ast.literal_eval)
```

```python

review_list = []
item_id_list = []
user_id_list = []
recommend_list = []


for lista in reviews['reviews']:
    for review in lista:
        review_list.append(review['review'])
        item_id_list.append(review['item_id'])
        user_id_list.append(reviews['user_id'])
        recommend_list.append(review['recommend'])
```

```python
review_singular = pd.DataFrame({
    'review': review_list,
    'item_id': item_id_list,
    'user_id': user_id_list,
    'recommend' : recommend_list
})
```

```python
review_singular.head()
```

Guardamos en archivo csv

```python
review_singular.to_csv('datasets/reviews_desanidado.csv')
```

Ahora haremos el sentiment analysis

```python
reviews = pd.read_csv('datasets/reviews_desanidado.csv')
reviews.drop(columns='Unnamed: 0', inplace=True)
reviews.head()
```

Importamos TextBlob

```python
from textblob import TextBlob
```

Aplicamos el sentiment analysis usando la funci√≥n polarity

```python
reviews['review'] = reviews['review'].astype(str)

# Define una funci√≥n para calcular la polaridad de una revisi√≥n
def calcular_polaridad(review):
    blob = TextBlob(review)
    return blob.sentiment.polarity

# Aplica la funci√≥n a la columna 'review' y crea la nueva columna 'polarity'
reviews['polarity'] = reviews['review'].apply(calcular_polaridad)

# Muestra el DataFrame con la nueva columna de polaridad
reviews.head()
```

Asignamos los valores 0, 1 y 2 dependiendo de la polaridad de la review

```python
def asignar_valor_sentimiento(polaridad):
    if polaridad < -0.2:
        return 0  # Mala
    elif -0.2 <= polaridad <= 0.2:
        return 1  # Neutral
    else:
        return 2  # Positiva

reviews['sentiment_value'] = reviews['polarity'].apply(asignar_valor_sentimiento)
```

```python
reviews.head()
```

Guardamos el dataset sin la columna review para facilitarlo al modelo de ML

```python
reviews_sentiment = reviews[['item_id', 'user_id', 'recommend', 'polarity', 'sentiment_value']]
reviews_sentiment.to_csv('datasets/reviews_sentiment.csv', index=False)
```

```python
reviews_sentiment.head()
```

# Dataset items

```python
items = pd.read_csv('datasets/Items.csv')
```

```python
items.head()
```

Borramos las columnas innecesarias

```python
items.drop(columns=['Unnamed: 0', 'user_url'], inplace=True)
```

```python
items.head()
```

Convertimos los valores de la columna items en lista

```python

def convertir_a_lista(valor):
    if isinstance(valor, str):
        try:
            return ast.literal_eval(valor)
        except (ValueError, SyntaxError):
            
            return valor
    else:
        return valor

items['items'] = items['items'].apply(convertir_a_lista)

items.head()
```

Desanidamos la lista de items

```python
items_desanidado = items.explode('items')

items_desanidado.head()
```

Borramos los registros nulos en items

```python
items_desanidado.dropna(subset='items', inplace=True)
```

Guardamos en archivo csv

```python
items_desanidado.to_csv('datasets/items_desanidado.csv', index=False)
```
